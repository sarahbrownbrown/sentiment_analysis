```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# load and clean data
avatar_data = pd.read_csv('avatar.csv', encoding = 'unicode_escape', engine='python')
avatar_lines = avatar_data.groupby('character').count()
avatar_lines = avatar_lines.sort_values(by=['character_words'], ascending=False)[:10]
top_character_names = avatar_lines.index.values

# get rid of non-top characters
character_sentiment = avatar_data[avatar_data['character'].isin(top_character_names)]
character_sentiment = character_sentiment[['character', 'character_words']]

# calculate the sentiment score
sent = SentimentIntensityAnalyzer()
character_sentiment.reset_index(inplace=True, drop=True)
character_sentiment[['negative', 'neutral', 'positive', 'compound']] = character_sentiment['character_words'].apply(sent.polarity_scores).apply(pd.Series)
character_sentiment
```
